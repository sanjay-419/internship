```python
import requests
import json
import pandas as pd
from google.colab import files
```

```python
df=pd.DataFrame()
response = requests.get("https://v6.exchangerate-api.com/v6/08866314444c246f31425991/latest/USD")
```

```python
if response.status_code==200:
    for i in range(1,400):
        response = requests.get("https://v6.exchangerate-api.com/v6/08866314444c246f31425991/latest/USD".format(i))
        t_df=pd.DataFrame(response.json())
        df=df.append(t_df,ignore_index=True)
else:
    print('error',response.status_code)
```
